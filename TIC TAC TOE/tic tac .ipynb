```python
def sum(a, b, c):
    return a + b + c

def printboard(xState, yState):
    for i in range(3):
        for j in range(3):
            index = i * 3 + j
            if xState[index]:
                print('X', end=' | ')
            elif yState[index]:
                print('O', end=' | ')
            else:
               print(index, end=' | ')
               print('\n---------')

def checkWin(xState, yState):
    wins = [[0, 1, 2], [3, 4, 5], [6, 7, 8], [0, 3, 6], [1, 4, 7], [2, 5, 8], [0, 4, 8], [2, 4, 6]]
    for win in wins:
        if sum(xState[win[0]], xState[win[1]], xState[win[2]]) == 3:
            print("X is the winner.")
            return 1
        if sum(yState[win[0]], yState[win[1]], yState[win[2]]) == 3:
            print("O is the winner.")
            return 0
    if all(xState[i] or yState[i] for i in range(9)):
        print("It's a draw!")
        return 2
    return -1

if __name__ == "__main__":  
    xState = [0] * 9
    yState = [0] * 9
    turn = 1
    print("Tic Tac Toe")
    while True:
        printboard(xState, yState)
        if turn == 1:
            print("X turn")
            while True:
                try:
                    value = int(input("Please enter your value (0-8): "))
                    if 0 <= value <= 8 and not xState[value] and not yState[value]:
                        xState[value] = 1
                        break
                    else:
                        print("Invalid move, try again.")
                except ValueError:
                    print("Invalid input, please enter a number.")
        else:
            print("O turn")
            while True:

                try:
                    value = int(input("Please enter your value (0-8): "))
                    if 0 <= value <= 8 and not xState[value] and not yState[value]:
                        yState[value] = 1
                        break
                    else:
                        print("Invalid move, try again.")
                except ValueError:
                    print("Invalid input, please enter a number.")
        cwin = checkWin(xState, yState)
        if cwin != -1:
            printboard(xState, yState)
            print("Match over")
            break
        turn = 1 - turn
```

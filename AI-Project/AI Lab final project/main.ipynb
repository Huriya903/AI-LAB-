```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack 
import pickle

df = pd.read_csv("q_quora.csv")

unnecessary_cols = ['Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12']
df = df.drop(columns=[col for col in unnecessary_cols if col in df.columns])
df['question1'] = df['question1'].fillna('')
df['question2'] = df['question2'].fillna('')
tfidf = TfidfVectorizer(max_features=10000)
q1_vec = tfidf.fit_transform(df['question1'])
q2_vec = tfidf.transform(df['question2'])

X = hstack([q1_vec, q2_vec])
y = df['is_duplicate'] 
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, shuffle=False)

model_svc = SVC()
model_svc.fit(train_X, train_y)
print("\nSVC Model successfully trained:")
print(model_svc)

with open('svc_model.pkl', 'wb') as file:
    pickle.dump(model_svc, file)
print("Model 'svc_model.pkl' mein save kar liya gaya hai.")
```
